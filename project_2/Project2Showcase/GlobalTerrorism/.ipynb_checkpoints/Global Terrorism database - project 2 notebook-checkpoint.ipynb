*** Exploring Global Terrorism ***

Throughout the years, global terrorism caused a lot of death and pain across the world. History has been marked with different events that will never be forgotten. Terrorists are becoming more creative and are continuously finding new ways to by pass security and cause maximum damage.
In this project, we will be exploring the "Global Terrorism Database" maintained by the National Consortium for the Study of Terrorism and Responses to Terrorism (START). The data contains 135 different columns and around 170K entries. It dates back to 1970 and covers all documented terrorist activities until 2016.The database itself has been updated by new attributes multiple on seperate occasion. All details are shared in the GTD Codebook available online on: https://www.start.umd.edu/gtd/downloads/Codebook.pdf

By exploring the data we are hoping to get more insights into:
1. Brief overview on terrorist attacks around the world:
    a)The deadliest attacks ever recorded
    b)The increase in terrorist attacks pre/post the arab spring
2. How terrorist attacks have shifted / increased from one region to another
3. Who is the most active terrorist group per period/years
4. How the most common methods of terrorism changed
5. Did the attacks become more massive in terms of # of victims in recent years compared to 1970th
6. How did the terrorist attacks changed by target (infrastructure, military, citizen)
7. How many terrorist were captured /sentence/killed vs. still at large per attack
8. Suicide attacks vs. non suicide
9. The first ever documented suicide jacket attack and how it spread across the world

Before starting our analysis, we would like to perform a quick sanity check on the data of the variables, though the data source is infamous for its high quality, we wouldn't be doing our due diligence if we just accept the data blindly. We will limit our data checking to a limited set of variables that we will be using in our analysis:
- "iyear": The data should begin in 1970 and end in 2016
- "nkill": The number of people killed cannot be negative
- "nwound": The number of people killed cannot be negative
- "region_txt": The region of where the attacks happened
- "attacktype1_txt": Attack type - Not to be confused with weapon subtype
- "weapsubtype1_txt": Subtype of weapon used during the attack

```python
# Data in the database is split ninto three columns. We want to add Date column and weekday for further analysis
file_copy=file[["eventid","iyear","imonth","iday"]]
file_copy=file_copy[file_copy.iday!=0]
file_copy=file_copy[file_copy.imonth!=0]
file_copy=file_copy[file_copy.iyear!=0]
file_copy["Date"]=pd.to_datetime(file_copy.iyear*10000+file_copy.imonth*100+file_copy.iday, format='%Y%m%d')
file_copy["Weekday"]=file_copy["Date"].dt.dayofweek
file=pd.concat([file, file_copy[["eventid", "Date", "Weekday"]]], axis=1, join='outer')
```

```python
# for high level time-series review we want to add column where years are split into groups
top_years_groups=[1970,1980,1990,2000,2010,2020]
year_groups=[str(x)+"th" for x in top_years_groups]
file['year_group']=file.iyear.apply(lambda a: str(next(x-10 for x in top_years_groups if x>a))+"th")
```

```python
#check completeness of data (# of empty rows) in the major columns we're going to use
nulls={}
for column in ['iyear','region_txt','country_txt','gname',\
               'nkill','nwound','attacktype1','weaptype1_txt','weapsubtype1_txt','suicide',\
              'targtype1_txt','latitude','longitude']:
    nulls[column]=len(file[file[column].isnull()==True])
    print(column,':',nulls[column])
```

```python
#Development of suicide attacks by regions, years, number of victims

top_groups=file.groupby('gname')['nkill'].sum().sort_values(ascending=False)[1:18]
suicide_regions=file[file.suicide==1].groupby("region_txt")['nkill'].sum()
suicide_regions=suicide_regions.sort_values(ascending=False)

suicide_groups=file[file.suicide==1].groupby("gname")['nkill'].sum()
suicide_groups=suicide_groups.sort_values(ascending=False)
suicide_groups.head()

colors = sns.color_palette("tab20", n_colors=20)
colors_gname={}
colors_short={}

for i, x in enumerate(list(suicide_groups[1:15].index)):
    colors_gname[x]=colors[i]
    colors_short[x]=colors[i]

for x in list(suicide_groups[15:].index):
    colors_gname[x]="grey"

colors_gname["Unknown"]="grey"
colors_short["Others"]="grey"
```

```python
ylabels=list(suicide_regions.index[:8])
ilen=len(ylabels)
pos=np.append(np.arange(0.5,2,0.8), np.arange(1.9,5,0.5))
    
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111)


for i, region in enumerate(ylabels):  
    x=file[(file.region_txt==region)&(file.suicide==1)].groupby(['iyear','gname'], as_index=False).agg({"nkill": "sum"})
    x=x[x.nkill>10]
    for j in range(len(x)):
        ax.scatter(x.iloc[j]['iyear'], pos[i], s=x.iloc[j]['nkill']*4, alpha=0.7, color=colors_gname[x.iloc[j]['gname']])
        
    
    
#x.iloc[k].iyear.INT-LOG
    
    
locsy, labelsy = plt.yticks(pos, ylabels)
plt.setp(labelsy, fontsize = 18)
#    ax.axis('tight')
ax.set_ylim(ymin = -0.1, ymax = ilen*0.5+0.5)
ax.set_xlim(xmin = 1980, xmax = 2016)
ax.grid(color = 'g', linestyle = ':')

markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors_short.values()]
ax.legend(markers, colors_short.keys(), numpoints=1, fontsize=18, loc='center left',  \
          bbox_to_anchor=(1.0, 0.5),  markerscale=3)

#formatter = DateFormatter("%d-%b '%y")
  
plt.xticks(fontsize=16) 

ax.invert_yaxis()
fig.autofmt_xdate()
ax.set_title("Spread of suicide attacks by regions", fontsize=24)
plt.show()
```

```python
#selecting the data I will use
regimes_short=regimes[['country','year','democ','autoc','polity','durable','byear']]
```

```python
#sanity check on names which mights be called differently in GTD
set(regimes_short.country.unique())-set(file.country_txt.unique())
```

```python
#correcting location names in accordance with GTD
regimes_short.Location=regimes_short.country.replace('Korea North','North Korea')
regimes_short.Location=regimes_short.country.replace('USSR','Soviet Union')
regimes_short.Location=regimes_short.country.replace('UAE','United Arab Emirates')
regimes_short.Location=regimes_short.country.replace('Vietnam North','Vietnam')
regimes_short.Location=regimes_short.country.replace('Korea South','South Korea')
regimes_short.Location=regimes_short.country.replace('Oman','Bosnia-Herzegovina')
regimes_short.Location=regimes_short.country.replace('Yemen North','Yemen')
regimes_short.Location=regimes_short.country.replace('Myanmar (Burma)','Myanmar')
regimes_short.Location=regimes_short.country.replace('Germany East','East Germany (GDR)')
```

```python
#add column to GTD on polity (political regime on the scale from -10 to 10)
def polity(x):
    try: return(regimes_short[(regimes_short.country==x['country_txt'])&\
                                         (regimes_short.year==x['iyear'])]['polity'].iloc[0])
    except: None
    

#file[]['democracy']=file.apply(democ, axis=1)
#file.iloc[1].apply(democ, axis=1)
polity(file.iloc[1])
file['polity']=file.apply(polity, axis=1)
```

```python
#add column to GTD on durability of the political regime in the country by that year
def durable(x):
    try: return(regimes_short[(regimes_short.country==x['country_txt'])&\
                                         (regimes_short.year==x['iyear'])]['durable'].iloc[0])
    except: None
    

#file[]['democracy']=file.apply(democ, axis=1)
#file.iloc[1].apply(democ, axis=1)
polity(file.iloc[1])
file['durable']=file.apply(durable, axis=1)
```

```python
#check whether terrorism depends on political regime in country

df=file[file.year_group=="2010th"].groupby(['polity'])['nkill'].sum()

fig, ax = plt.subplots(figsize=(15,8))

plot=df.plot.barh(label=polity, color=colors)
plot.tick_params(axis='both', which='major', labelsize=14)
plot.set_title("Fatality of terrorism in 2010th, depending on political regime", fontsize = 18)
#plt.legend(fontsize=14, loc='center left',  bbox_to_anchor=(1.0, 0.5))
```

```python
#check whether terrorism depends on political transitions period, power shifts and durability of existing regime
df=file.groupby(['durable'])['nkill'].sum().head(40)

fig, ax = plt.subplots(figsize=(18,8))

plot=df.plot.bar(label="years", color=colors, width=1)
plot.tick_params(axis='both', which='major', labelsize=14)
plot.set_title("Fatality from terrorist acts, depending on longevity of political regime", fontsize = 18)
plt.xlabel('Political regime, years', fontsize=14)
plt.ylabel('Total # of fatalities', fontsize=14)
```

```python
#Detailed review on how terrorism corresponds with political instability and shifts of power. 
#Chart for countries with biggest number of terrorism fatalities 


top_countries=file.groupby('country_txt')['nkill'].sum().sort_values(ascending=False)[:15]
ylabels=list(top_countries.index)
ilen=len(ylabels)
pos=np.append(np.arange(0.5,4,0.8),np.arange(4.2,4.2+0.5*10,0.5))
    
fig = plt.figure(figsize=(20,25))
ax = fig.add_subplot(111)


for i, country in enumerate(ylabels):  
    x=regimes_short[(regimes_short.country==country)&(regimes_short.durable==0)]
    for j in range(len(x)):
        ax.plot(x.iloc[j].year, pos[i], marker="D", linestyle="", markersize=10, alpha=1.0, color="blue")
    y=file[file.country_txt==country].groupby('iyear')['nkill'].sum()
    for k in list(y.index):  
        ax.scatter(k, pos[i],s=y[k], alpha=0.8)

    
#x.iloc[k].iyear.INT-LOG
    
    
locsy, labelsy = plt.yticks(pos, ylabels)
plt.setp(labelsy, fontsize = 14)
#    ax.axis('tight')
ax.set_ylim(ymin = -0.1, ymax = ilen*0.5+0.5)
ax.set_xlim(xmin = 1970, xmax = 2016)
ax.grid(color = 'g', linestyle = ':')


#formatter = DateFormatter("%d-%b '%y")
  
plt.xticks(fontsize=14) 

ax.invert_yaxis()
fig.autofmt_xdate()
ax.set_title("Change of political regime and terrorism acts", fontsize=20)
plt.show()
```
